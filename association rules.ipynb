```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_excel('Online retail.xlsx',header=None)
df
```

```python
dg=df[0].str.split(',', expand=True)
```

```python
dg
```

```python
dg[0].value_counts()
```

```python
dg[1].value_counts()
```

```python
dg=pd.get_dummies(dg)
```

```python
dg
```

```python
## apriori algorithum
```

```python
frequent_items=apriori(dg,min_support=0.001,use_colnames=True)
frequent_items
```

```python
dg
```

```python
frequent_items=apriori(dg,min_support=0.01,use_colnames=True)
frequent_items
```

```python
frequent_items.info()
```

```python
rules=association_rules(frequent_items,metric='lift',min_threshold=1.0)
rules
```

```python
rules[['confidence','support','antecedent support']].hist(figsize=(10,6))
```

```python
rules[['consequent support','lift']].hist(figsize=(10,6))
```

```python
plt.scatter(rules['confidence'],rules['support'])
```

```python
plt.scatter(rules['lift'],rules['support'])
```
