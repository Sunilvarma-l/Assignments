```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('wine.csv')
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaled_df=scaler.fit_transform(df)
```

```python
pca=PCA(n_components=12)
pca.values=pca.fit_transform(scaled_df)
```

```python
pca.components_
```

```python
d=pd.DataFrame(pca.components_)
```

```python
d
```

```python
var=pca.explained_variance_ratio_
```

```python
var
```

```python
var1=np.cumsum(np.round(var,decimals=4)*100)
```

```python
var1
```

```python
plt.plot(var1,color='red')
```

```python
pca.values[:,0:2]
```

```python
finalidf=pd.DataFrame(pca.values[:,0:2],columns=['pca1','pca2'])
```

```python
finalidf
```

```python
plt.scatter(x='pca1',y='pca2',data=finalidf,s=100)
```

```python
d
```

```python
import scipy .cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
dendrogram=sch.dendrogram(sch.linkage(finalidf,method='ward'))
```

```python
hc=AgglomerativeClustering(n_clusters=3,metric='euclidean',linkage='average')
```

```python
hc
```

```python
y_hc=hc.fit_predict(finalidf)
```

```python
y_hc
```

```python
finalidf['clusters']=y_hc
```
